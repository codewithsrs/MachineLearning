```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', 100)
play_data_train = pd.read_csv("/kaggle/input/playground-series-s3e13/train.csv")
play_data_train = play_data_train.sample(frac = 1)
play_data_test = pd.read_csv("/kaggle/input/playground-series-s3e13/test.csv")
play_data_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
play_data_train['prognosis'] = lb.fit_transform(play_data_train['prognosis'])
```

```python
play_data_train.head()
```

```python
play_data_train.drop(['bullseye_rash','speech_problem','toenail_loss','ulcers','lips_irritation'
                      ,'itchiness','finger_inflammation','toe_inflammation','microcephaly',
                        'rigor','bitter_tongue','convulsion','anemia','cocacola_urine','hypoglycemia'
                      ,'prostraction','hyperpyrexia',
                        'stiff_neck','irritability','confusion','tremor','paralysis','lymph_swells'
                      ,'breathing_restriction'], axis = 1, inplace = True)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = keras.Sequential(
    [
        layers.Dense(540, activation="relu", name="layer1"),
        keras.layers.Dropout(0.6),
        layers.Dense(540, activation="relu", name="layer2"),
        keras.layers.Dropout(0.4),
        layers.Dense(540, activation="relu", name="layer3"),
        keras.layers.Dropout(0.5),
        layers.Dense(256, activation="relu", name="layer4"),
        keras.layers.Dropout(0.6),
        layers.Dense(256, activation="relu", name="layer5"),
        keras.layers.Dropout(0.6),
        layers.Dense(256, activation="relu", name="layer6"),
        keras.layers.Dropout(0.6),
        layers.Dense(11, activation="softmax", name="layer7"),
    ]
)
model.compile(optimizer = tf.optimizers.Adam(),
              loss = 'sparse_categorical_crossentropy',
              metrics =['accuracy'])
```

```python
play_data_train.drop('id', axis = 1, inplace = True)
x = play_data_train.drop('prognosis', axis = 1)[:350]
y = play_data_train['prognosis'][:350]
```

```python
x_val = play_data_train.drop('prognosis', axis = 1)[350:]
y_val = play_data_train['prognosis'][350:]
```

```python
for i in range(5):
    model.fit(x,y, epochs = 20)
```

```python
model.evaluate(x_val, y_val)
```

```python
play_data_test.drop('id', axis = 1, inplace = True)
```

```python
play_data_test.drop(['bullseye_rash','speech_problem','toenail_loss','ulcers','lips_irritation'
                      ,'itchiness','finger_inflammation','toe_inflammation','microcephaly',
                        'rigor','bitter_tongue','convulsion','anemia','cocacola_urine','hypoglycemia'
                      ,'prostraction','hyperpyrexia',
                        'stiff_neck','irritability','confusion','tremor','paralysis','lymph_swells'
                      ,'breathing_restriction'], axis = 1, inplace = True)
```

```python
pred = model.predict(play_data_test)
```

```python
prediction = []
for i in pred:
    prediction.append(np.argmax(i))
```

```python
submission = pd.read_csv("/kaggle/input/playground-series-s3e13/sample_submission.csv")
submission.head()
```

```python
prediction = lb.inverse_transform(prediction)
```

```python
submission['prognosis'] = prediction
```

```python
submission.to_csv('submission.csv', index = False)
```
